### Aggregate

```python
registrations.groupby(registrations.Make).size().sort_values(ascending = False)
```

```python
registrations.groupby(['Make'])['Electric Range'].aggregate('mean').sort_values(ascending = False)
```

```python
registrations[registrations['Transaction Year'] == 2019 ].groupby(['Model']).size().sort_values(ascending = False)
```

```python
total_reg = registrations.groupby('Transaction Year').size().sum()
total_reg
```

### Plot

```python
plt.rcParams['figure.figsize'] = [15, 6] # Taille du graphique en pouces
plt.rcParams['figure.dpi'] = 200 # résolution en points par pouce
registrations.groupby('Transaction Year').size().plot()
```

```python
plt.plot(registrations[(registrations.Make == "TESLA")].groupby("Transaction Year").size(), label = 'Tesla')
plt.plot(registrations[(registrations.Make == "NISSAN")].groupby("Transaction Year").size(), label = 'Nissan')
plt.plot(registrations[(registrations.Make == "CHEVROLET")].groupby("Transaction Year").size(), label = 'Chevrolet')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)
plt.show()
```

```python
cars_perY_perMake = registrations.pivot_table(values='Model', index = 'Transaction Year', columns = 'Make', aggfunc = 'count')
cars_perY_perMake
```

```python
cars_perY_perMake.plot(title = "Cars per year per make") 
```
